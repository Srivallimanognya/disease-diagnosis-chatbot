```python
import pandas as pd

df = pd.read_csv("heart.csv")   # ‚Üê I will replace with your filename
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

X = df.drop("target", axis=1)
y = df["target"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model = LogisticRegression()
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("Model Accuracy:", accuracy)
```

```python


def heart_chatbot():
    print("üíì Welcome to the Heart Disease Diagnosis Chatbot!")
    print("Please answer the medical questions below:\n")

    # Collect inputs
    age = float(input("Age: "))
    sex = float(input("Sex (1=Male, 0=Female): "))
    cp = float(input("Chest Pain Type (0‚Äì3): "))
    trestbps = float(input("Resting Blood Pressure: "))
    chol = float(input("Cholesterol Level: "))
    fbs = float(input("Fasting Blood Sugar >120 (1/0): "))
    restecg = float(input("Resting ECG (0‚Äì2): "))
    thalach = float(input("Max Heart Rate Achieved: "))
    exang = float(input("Exercise Induced Angina (1/0): "))
    oldpeak = float(input("ST Depression Value: "))
    slope = float(input("Slope (0‚Äì2): "))
    ca = float(input("Major Vessels (0‚Äì3): "))
    thal = float(input("Thalassemia (1,2,3): "))

    # Convert to DataFrame with feature names
    user_df = pd.DataFrame([{
        "age": age,
        "sex": sex,
        "cp": cp,
        "trestbps": trestbps,
        "chol": chol,
        "fbs": fbs,
        "restecg": restecg,
        "thalach": thalach,
        "exang": exang,
        "oldpeak": oldpeak,
        "slope": slope,
        "ca": ca,
        "thal": thal
    }])

    # Scale the input using the fitted scaler
    user_scaled = scaler.transform(user_df)

    # Make prediction
    prediction = model.predict(user_scaled)

    print("\nüìä **Diagnosis Result**:")
    if prediction[0] == 1:
        print("‚ö†Ô∏è High risk of heart disease. Please consult a cardiologist.")
    else:
        print("‚úÖ Low risk of heart disease. No major symptoms detected.")


# Run the chatbot
heart_chatbot()
```
